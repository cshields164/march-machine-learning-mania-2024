```python
# We are going to import some Libraries that we are going to use in the project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# We are going to import our dataset form multiple csv files
df1 = pd.read_csv('C:/Users/Alex/Downloads/march-machine-learning-mania-2024/MTeams.csv')
df2= pd.read_csv('C:/Users/Alex/Downloads/march-machine-learning-mania-2024/MSeasons.csv')
df3= pd.read_csv('C:/Users/Alex/Downloads/march-machine-learning-mania-2024/MNCAATourneySeeds.csv')
df4= pd.read_csv('C:/Users/Alex/Downloads/march-machine-learning-mania-2024/MRegularSeasonCompactResults.csv')
df5 = pd.read_csv('C:/Users/Alex/Downloads/march-machine-learning-mania-2024/MNCAATourneyCompactResults.csv')
```

```python
# We are going to merge the first 2 datasets to see what we get and how we can use it or if we need to clean it
df1 = df1.merge(df2, left_on='FirstD1Season', right_on='Season', how='left')
df1.head()
```

```python
# To keep the dataset clean we are going to start to drop some columns that we are not going to use
# We are dropping 'FirstD1Season', 'LastD1Season','RegionW','RegionX','RegionY','RegionZ','DayZero'
# Because we feel like this is erlevant to our analysis, We are also going to keep checking our dataset to see if we need to drop more columns
df1.drop(['FirstD1Season', 'LastD1Season','RegionW','RegionX','RegionY','RegionZ','DayZero'], axis=1, inplace=True)
df1.head()
```

```python
df1 = df1.merge(df3, left_on='TeamID', right_on='TeamID', how='left')
df1.head()
```

```python
df1.drop(['Season_x', 'Season_y'], axis=1, inplace=True)
df1.head()
```

```python
merged = df1.merge(df4, left_on='TeamID', right_on='WTeamID', how='inner')
```
